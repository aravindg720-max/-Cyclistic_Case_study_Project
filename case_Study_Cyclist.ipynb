```python
import pandas as pd
import glob
```

```python
path = r'C:\Users\91894\Desktop\Data Analytics-Road -Map\FInal-Projects\Case-study Cyclistic Project #2\Dataset cyclist'
all_files =glob.glob(path +"\*.csv")

li = []
for filename in all_files:
    df=pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
merged_df = pd.concat(li, axis=0, ignore_index=True)
print(f"Merged Successfully {len(all_files)} files")
```

```python
merged_df['started_at'] = pd.to_datetime(merged_df['started_at'])
merged_df['ended_at'] = pd.to_datetime(merged_df['ended_at'])
```

```python
merged_df['ride_length'] = merged_df['ended_at'] - merged_df['started_at']
```

```python
merged_df.head()
```

```python
merged_df['day_of_week'] =merged_df['started_at'].dt.weekday
merged_df['day_of_week'] =merged_df['day_of_week'].map({0:2,1:3,2:4,3:5,4:6,5:0,6:1})
```

```python
merged_df.head()
```

```python
merged_df= merged_df[merged_df['ride_length']>pd.Timedelta(minutes=1)]
merged_df.head()
```

```python
test_filter = (
    merged_df['start_station_name'].str.contains('test|base', case=False, na=False) |
    merged_df['end_station_name'].str.contains('test|base', case=False, na=False) |
    merged_df['start_station_id'].str.contains('test', case=False, na=False)
)
merged_df = merged_df[~test_filter]
```

```python
merged_df =merged_df.drop_duplicates(subset=['ride_id'])
print(f"Total duplicate rows removed: {len(merged_df) - len(merged_df.drop_duplicates(subset=['ride_id']))}")
```

```python
station_cols = ['start_station_name', 'end_station_name', 'start_station_id', 'end_station_id']
for col in station_cols:
   merged_df[col] = merged_df[col].fillna('Unknown')
print("Missing values in station columns have been filled with 'Unknown' the columns are:")
```

```python
merged_df = merged_df.dropna(subset=['end_lat', 'end_lng'])
```

```python
merged_df.isnull().sum()
```

```python
merged_df=merged_df[merged_df['ride_length']>pd.Timedelta(minutes=1)]
merged_df.head()
```

```python
merged_df['ride_length_in_seconds']= merged_df['ride_length'].dt.total_seconds()
merged_df.head()
```

```python
merged_df =merged_df.drop(columns=['ride_length'])
merged_df.head()
```

```python

merged_df['ride_length_minutes'] = merged_df['ride_length_in_seconds'] / 60
```

```python
merged_df['month']= merged_df['started_at'].dt.month_name()
merged_df['hour']= merged_df['started_at'].dt.hour
merged_df.head()
```

```python
from sqlalchemy import create_engine

username = "root"          # your MySQL username
password = "admin"         # your MySQL password
host = "localhost"
port = "3306"
database = "cyclist"  # your MySQL database name

engine = create_engine(f'mysql+pymysql://{username}:{password}@{host}:{port}/{database}')

table_name = 'cyclist_rides'  # name of the table to store the data
merged_df.to_sql(table_name, con=engine, if_exists='replace', index=False)

print("DataFrame has been successfully written to the MySQL database.")
```
